# There Interfaces
>- **Pylab (deprecated)**<br>
>- **Pyplot (scripting interface)**<br>
>- **Object oriented API**

## Pylab
>- **Pros**<br>
        . Matlab-like interface
        . Less verbose
>- **Cons**<br>
        . import everything into global space
        . More obscure code
>- **When to use it**<br>
        . One time usage

## Pyplot
>- **Pros**<br>
        . Less obscure code
        . Namespaced
>- **Cons**<br>
        . Less controlled 
>- **When to use it**<br>
        . Interactive Data Exploration (EDA)
        . Small batch vizs

## Object Oriented API
>- **Pros**<br>
        . Total control (more towards s/w engg)
>- **Cons**<br>
        . Syntactically heavy 
>- **When to use it**<br>
        . When embedding
        . Large batch vizs

```python
# this applies to people using retina screens only 
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.random.randn(1000) # 
y = np.random.randn(1000)
```

```python
'''
plt.figure is the class that holds all figure objects in matplot lib.
It checks if there is a figure object if not creates a new object
'''
plt.figure(figsize=(5,5)) 
plt.scatter(x,y)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.title("Normally distributed data with $\mu =0, \sigma =1$")
```
